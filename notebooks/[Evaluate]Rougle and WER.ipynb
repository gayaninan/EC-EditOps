```python
def calculate_wer( col1, col2, path_):
    df = pd.read_csv(path_)
    trans_batch = df[col1].tolist()
    refs_batch = df[col2].tolist()
    score = wer_metric.compute(predictions=refs_batch, references=trans_batch)
    return score

def calculate_rouge( col1, col2, path_):
    df = pd.read_csv(path_)
    df = df.dropna()
    predictions = df[col1].tolist()
    references = df[col2].tolist()
    results = rouge_metric.compute(predictions=predictions, references=references)
    return results

def generate_results(input_path, output_path):
    baseline_col1 = 'refs'
    baseline_col2 = 'trans'

    col1 = 'refs'
    col2 = 'model_corrected'

    baseline_wer = calculate_wer(baseline_col1, baseline_col2, input_path)
    baseline_rouge = calculate_rouge(baseline_col1, baseline_col2, input_path)

    wer = calculate_wer(col1, col2, input_path)
    rouge = calculate_rouge(col1, col2, input_path)

    with open(output_path, 'w') as text_file:
        text_file.write(output_path)
        text_file.write('Baseline WER: %s\n' % baseline_wer)
        text_file.write('WER: %s\n' % wer)
        text_file.write('Baseline ROUGE: {}\n'.format(baseline_rouge))
        text_file.write('ROUGE: {}\n'.format(rouge))
```

```python
wer_metric = load_metric('wer')
rouge_metric = load_metric('rouge')
```
